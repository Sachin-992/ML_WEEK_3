 Performance Comparison Report 

```python
print("PERFORMANCE SUMMARY \n")

def summarize_performance(name, history):
    best_acc = max(history['accuracy']) * 100
    best_val_acc = max(history['val_accuracy']) * 100
    best_loss = min(history['loss'])
    best_val_loss = min(history['val_loss'])
    
    print(f"   {name}")
    print(f"   Max Training Accuracy:     {best_acc:.2f}%")
    print(f"   Max Validation Accuracy:   {best_val_acc:.2f}%")
    print(f"   Min Training Loss:         {best_loss:.4f}")
    print(f"   Min Validation Loss:       {best_val_loss:.4f}\n")

summarize_performance("Basic CNN", cnn_history)
summarize_performance("ResNet", resnet_history)
summarize_performance("ResNet + Augmentation", resnet_aug_history)
```
